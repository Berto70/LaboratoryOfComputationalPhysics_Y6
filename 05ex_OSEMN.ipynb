4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
# credit_card = np.genfromtxt('./credit_card.dat')

with  open('credit_card.dat','rb') as f:
    data = f.readlines()
    
for row in data:
    for d in range(int((len(row)-4)/6)):
        print(chr(int(row[d*6:d*6+6], 2)), end='')
    print('')
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json
data_json = json.load(open('user_data.json'))

# Use lambda function to filter the data
filtered_data = list(filter(lambda credit: credit['CreditCardType'] == "American Express", data_json))

# Print the filtered data
print(json.dumps(filtered_data, indent=2))

with open("./05ex/user_data_filtered.csv", 'w') as f:
    f.write("ID,JobTitle,EmailAddress,FirstNameLastName,CreditCard,CreditCardType\n")
    for row in filtered_data:
        f.write(row['ID'] + "," + row['JobTitle'] + "," + row['EmailAddress'] + "," + row['FirstNameLastName'] + "," + row['CreditCard'] + ',' +row['CreditCardType'] + "\n")
```

7\. Load the remote file [https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1](https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1) with Pandas and plot a scatter plot all possible combination of the following fields:
    
  + features_1
  + features_2
  + features_3

```python
regr_data = pd.read_csv("./05ex/regression_generated.csv")

feat = ['features_1', 'features_2', 'features_3']

fig, ax = plt.subplots(3, 3, figsize=(15, 15))

for i in range(3):
    for j in range(3):
        ax[i,j].scatter(regr_data[feat[i]], regr_data[feat[j]])
        ax[i,i].scatter(regr_data[feat[i]], regr_data[feat[i]], color='orange')
        ax[i,j].set(xlabel=feat[i], ylabel=feat[j])

plt.show()
```
